```python
import pandas as pd
import os
import glob

metrics_folder = "../outputs/metrics"

all_dfs = []

# Read all metrics files
for file in glob.glob(f"{metrics_folder}/*.csv"):

    # Skip previously generated summary files
    if "final_all_experiments" in file:
        continue

    df = pd.read_csv(file)

    experiment_name = os.path.basename(file)
    experiment_name = experiment_name.replace("_comparison_metrics.csv", "")
    experiment_name = experiment_name.replace(".csv", "")

    df["Feature_Set"] = experiment_name

    all_dfs.append(df)

# Combine all experiments
combined_metrics = pd.concat(
    all_dfs,
    ignore_index=True
)

# Convert to wide format
wide_df = combined_metrics.pivot(
    index="Feature_Set",
    columns="Model",
    values=[
        "MAE",
        "MedAE",
        "RMSE",
        "R2",
        "Pearson_r",
        "Spearman_rho"
    ]
)

# Flatten column names
wide_df.columns = [
    f"{metric}_{model.replace(' Regression', '')}"
    for metric, model in wide_df.columns
]

# Convert index back to column
wide_df = wide_df.reset_index()

# Save final file
wide_df.to_csv(
    "../outputs/final_comparison.csv",
    index=False
)

print("Saved Successfully!")

wide_df
```
